# FORMATO .ZIP


```python
import gzip
import zipfile
```

```python
def descargar_datos_sst(anios, meses, dias, nombre_archivo, carpeta_destino):
    for idx in range(len(anios)):
        c = cdsapi.Client()
        try: # Lo que espero que ocurra
            c.retrieve(
                'satellite-sea-surface-temperature',
                {
                    'version': '2_1',
                    'variable': 'all',
                    'format': 'zip',
                    'processinglevel': 'level_4',
                    'sensor_on_satellite': 'combined_product',
                    'year': str(anios[idx]),
                    'month': meses,
                    'day': dias,
                },
                nombre_archivo)

            with zipfile.ZipFile(nombre_archivo, 'r') as zip_ref:
                zip_ref.extractall(carpeta_destino)

        except Exception as e: # Las excepciones
            print(f"Error al descargar o procesar datos: {str(e)}")
```

```python
# Ejemplo de uso de la función
anios = [2021, 2022]
meses = ['01','03']
dias = ['30', '31']

nombre_archivo = 'datos_sst.zip'
carpeta_destino = 'datos_sst'
# Ok:
descargar_datos_sst(anios, meses, dias, nombre_archivo, carpeta_destino)
```
